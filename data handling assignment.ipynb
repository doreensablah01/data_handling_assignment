```python
#question 10a: visualizing data by generating heatmap before eliminating NaN values. 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("data_handling.csv")
data_1 = data.loc[:,['PassengerId','Survived', 'Pclass', 'Age','SibSp','Parch']]
print(data_1)
sns.heatmap(data_1, cmap='BuPu',linecolor = 'black')
plt.title('Titanic Death Plot')
plt.show()
```

```python
#question 10b: visualizing data by generating heatmap after eliminating NaN values.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("data_handling.csv")
data.dropna(inplace = True)
data_1 = data.loc[:,['PassengerId','Survived', 'Pclass', 'Age','SibSp','Parch']]
print(data_1)
sns.heatmap(data_1, linewidth = .5, cmap='coolwarm')
plt.title('Titanic Death Plot')
plt.show()
```

```python
#question 11: determining the mean, median and mode for all features in data.
print('MEAN')
print('PassengerId: ' + str(data['PassengerId'].mean()))
print('Survived: ' + str(data['Survived'].mean()))
print('Pclass: ' + str(data['Pclass'].mean()))
print('Age: ' + str(data['Age'].mean()))
print('SibSp: ' + str(data['SibSp'].mean()))
print('Parch: ' + str(data['Parch'].mean()))
print('Fare: ' + str(data['Fare'].mean()))
print('******************************************************')
print('MEDIAN')
print('PassengerId: ' + str(data['PassengerId'].median()))
print('Survived: ' + str(data['Survived'].median()))
print('Pclass: ' + str(data['Pclass'].median()))
print('Age: ' + str(data['Age'].median()))
print('SibSp: ' + str(data['SibSp'].median()))
print('Parch: ' + str(data['Parch'].median()))
print('Fare: ' + str(data['Fare'].median()))
print('******************************************************')
print('MODE')
print(data['Survived'].mode())
print(data['Pclass'].mode())
print(data['Age'].mode())
print(data['SibSp'].mode())
print(data['Parch'].mode())
print(data['Fare'].mode())
```

```python
#question 13: performing a bivariate analysis using scatterplot 
#to show the relationship between any two features of the dataset
#barchart
import os
os.getcwd()
os.chdir('C:\\Users\\user\\Desktop\\awo')
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("data_handling.csv")
data.plot(x ='Age', y='Fare',kind = 'scatter', s=30, marker = '+', color = 'r', label = 'Age/Fare')
plt.xlabel('Age') 
plt.ylabel('Fare')
plt.title('The distribution of passengers across sex, passenger class and Embarked')
plt.legend() 
plt.show()
```
